Vetores em $\mathbb{R}^2$ ou $\mathbb{R}^3$
===========================================

O conjunto $\mathbb{R}^2=\mathbb{R}\times\mathbb{R}$ será chamado de
**plano euclidiano** ou de **espaço euclidiano bidimensional** e seus
elementos serão chamados de **vetores bidimensionais**. De forma análoga
o conjunto $\mathbb{R}^3=\mathbb{R}\times\mathbb{R}\times\mathbb{R}$
será chamado de **espaço euclidiano tridimensional** e seus elementos
serão chamados de **vetores tridimensionais**. Em alguns casos os
elementos de $\mathbb{R}^2$ e $\mathbb{R}^3$ serão chamados simplesmente
de **pontos**.

Os elementos do plano $\mathbb{R}^2$ são os pares ordenados $(x,y)$, em
que $x,y\in\mathbb{R}$; enquanto que os elementos do espaço
$\mathbb{R}^3$ são as ternas ordenadas $(x,y,z)$, em que
$x,y,z\in\mathbb{R}$.

Dados dois vetores bidimensionais $u=(a,b)$ e $v=(c,d)$, tem-se que
$u=v$ se, e somente se, $a=c$ e $b=d$. Uma afirmação análoga vale no
caso de vetores tridimensionais.

Dado um vetor $v=(x,y)\in\mathbb{R}^2$, o número $x$ é chamado de
**primeira componente** ou **primeira coordenada** ou **componente
$\boldsymbol{x}$** do vetor $v$. De forma análoga, o número $y$ é
chamado de **segunda componente** ou **segunda coordenada** ou
**componente $\boldsymbol{y}$** de $v$. Definições análogas podem ser
elaboradas para vetores em $\mathbb{R}^3$.

Exemplo: Os vetores $(2,1)$ e $(2,3)$ são diferentes pois $1\ne 3$. Da
mesma forma, os vetores $(1,5)$ e $(5,1)$ são diferentes pois $1\ne 5$.

Vetores de $\mathbb{R}^2$ e $\mathbb{R}^3$ são usualmente representados
graficamente por **setas** que têm comprimento e direção definidos mas
cuja localização da sua origem é irrelevante. Definidos os eixos
retangulares $x$ e $y$ no plano, para representar graficamente um vetor
$v=(\alpha,\beta)$ primeiramente escolhemos um ponto qualquer $P_0$ do
plano como a origem do vetor $v$. Para determinar a posição da ponta do
vetor $v$, partindo de $P_0$, move-se $\alpha$ unidades na direção do
eixo $x$ e depois $\beta$ unidades na direção do eixo $y$, chegando em
um ponto $P$. Assim, o vetor $v$ será representado pela seta que tem
origem $P_0$ e cuja ponta está no ponto $P$. Pode-se perceber facilmente
que se o ponto $P_0$ tem coordenadas $(\alpha_0,\beta_0)$, o ponto $P$
tem coordenadas $(\alpha_0+\alpha,\beta_0+\beta)$. Além disso, o
comprimento da seta obtida pode ser calculada usando o teorema de
Pitágoras resultando em $\sqrt{\alpha^2+\beta^2}$. A representação
gráfica de um vetor em $\mathbb{R}^3$ pode ser construída de forma
semelhante.

Define-se a **norma** de um vetor $u=(x,y)\in\mathbb{R}^2$ por
$|u|=\sqrt{x^2+y^2}$. A norma de um vetor $v=(x,y,z)\in\mathbb{R}^3$ é
definida por $|v|=\sqrt{x^2+y^2+z^2}$. Segue dessa definição que a norma
de um vetor é igual ao comprimento da seta que o representa.

Operações básicas com vetores
=============================

Dados os vetores $u=(x_1,y_1)$ e $v=(x_2,y_2)$ e o número
$\alpha\in\mathbb{R}$, definem-se as operações de **adição de vetores**
e **multiplicação por um número** pelas seguintes igualdades:
$$\begin{split}
u+v&=(x_1+x_2,y_1+y_2)\\
\alpha v&=(\alpha x_2,\alpha y_2)\,.
\end{split}$$ Pode-se verificar facilmente que essas operações têm as
seguintes propriedades:

1.  associatividade: $u+(v+w)=(u+v)+w$ e
    $\alpha(\beta v)=(\alpha\beta)v$;

2.  comutatividade: $u+v=v+u$;

3.  vetor nulo: existe o vetor $0=(0,0)\in\mathbb{R}^2$ tal que $v+0=v$
    para todo $v\in\mathbb{R}^2$;

4.  inverso aditivo: para todo $v\in\mathbb{R}^2$, existe
    $-v\in\mathbb{R}^2$ tal que $v+(-v)=0$;

5.  multiplicação por $1$: $1v=v$;

6.  distributividade: $(\alpha+\beta)v=\alpha v+\beta v$ e
    $\alpha(u+v)=\alpha u+\alpha v$.

As operações de adição e multiplicação por um número no espaço
$\mathbb{R}^3$ podem ser definidas de forma análoga. Essas operações
também têm as mesmas propriedades que suas análogas no espaço
$\mathbb{R}^2$.

A soma de dois vetores $u$ e $v$, ambos em $\mathbb{R}^2$ ou em
$\mathbb{R}^3$, pode ser obtida graficamente colocando a origem do vetor
$v$ na ponta do vetor $u$ e assim o vetor $u+v$ será representado pela
seta cuja origem é a do vetor $u$ e cuja ponta é a do vetor $v$. Por
outro lado, a multiplicação de um número $\alpha>0$ ($\alpha<0$) pelo
vetor $v$ será representado graficamente por uma seta na mesma direção
(direção contrária) do vetor $v$ e cujo comprimento é $\alpha$ vezes o
dele.

Exemplo: Sejam $u=(a,b)$ e $v=(c,d)$ vetores não nulos. Tem-se que $u$ é
um múltiplo de $v$ (graficamente, $u$ paralelo a $v$) se, e somente se,
$ad-bc=0$. Com efeito, se $u=\alpha v$, $a=\alpha c$ e $b=\alpha d$ e,
por conseguinte, $ad=bc$. Por outro lado, se $ad-bc=0$ e $a\ne 0$, então
$d=bc/a$. Segue daqui que $c\ne 0$, pois $c=0$ implicaria que $d=0$ e
assim teríamos que $v=0$. Pondo $\alpha=c/a$, vemos que $d=\alpha b$ e
$c=\alpha a$.

Dependência e independência linear
==================================

Uma **combinação linear** dos vetores $u_1,\ldots,u_n$, todos em
$\mathbb{R}^2$ ou $\mathbb{R}^3$, é um vetor
$v=\alpha_1u_1+\cdots+\alpha_n u_n$, em que
$\alpha_1,\ldots,\alpha_n\in\mathbb{R}$.

Seja $X$ um conjunto de vetores em $\mathbb{R}^2$ ou em $\mathbb{R}^3$.
Diz-se que $X$ é um conjunto **linearmente dependente** (**L.D.**) se
existe um vetor $v\in X$ que pode ser escrito como uma combinação linear
de vetores de $X$ diferente de $v=1v$.

Exemplos:

1.  O conjunto
    $X=\{(1,2,1),(-1,2,1),(0,4,2),(0,0,1)\}\subset\mathbb{R}^2$ é um
    conjunto L.D., pois $(2,4,2)=(1,2,1)+(-1,2,1)$.

2.  Qualquer subconjunto $X$ de $\mathbb{R}^2$ ou de $\mathbb{R}^3$ que
    contém o vetor nulo é L.D., pois o vetor nulo pode ser escrito como
    uma combinação linear de vetores de $X$ com coeficientes nulos.

3.  Todo conjunto de vetores que contem um subconjunto L.D. é também
    L.D..

Diz-se que um conjunto $X$ em $\mathbb{R}^2$ ou em $\mathbb{R}^3$ é
**linearmente independente** (**L.I.**) se não é L.D..

**Teorema.** Um conjunto $X=\{u_1,\ldots,u_n\}$ em $\mathbb{R}^2$ ou em
$\mathbb{R}^3$ é L.I. se, e somente se, a condição
$\alpha_1u_1+\cdots+\alpha_nu_n=0$ implica que
$\alpha_1=\ldots=\alpha_n=0$.

*Demonstração.* ($\Rightarrow$) Se existissem coeficientes
$\alpha_1,\ldots,\alpha_n$ não todos nulos tais que
$\alpha_1u_1+\cdots+\alpha_nu_n=0$, após uma possível renomeação dos
vetores podemos ter $\alpha_1\ne 0$. Logo,
$$u_1=-\frac{\alpha_2}{\alpha_1}u_2-\cdots-\frac{\alpha_n}{\alpha_1}u_n\,,$$
ou seja, $u_1$ seria uma combinação linear dos outros vetores de $X$.
Portanto, $X$ seria L.D.. ($\Leftarrow$) Se $X$ fosse L.D., após uma
possível renomeação dos vetores podemos ter
$u_1=\alpha_2u_2+\cdots+\alpha_nu_n$. Logo,
$u_1-\alpha_2u_2-\cdots-\alpha_nu_n=0$ e nessa igualdade, pelo menos o
coeficiente de $u_1$ é diferente de $0$. $\square$

Exemplos:

1.  O conjunto $\{(1,2,1),(1,0,0)\}\subset\mathbb{R}^3$ é L.I.. Com
    efeito, se $$\alpha(1,2,1)+\beta(1,0,0)=(0,0,0)\,,$$ então
    $$\begin{split}
    \alpha+\beta&=0\\
    2\alpha&=0\\
    \alpha&=0\,.
    \end{split}$$ Portanto, $\alpha=0$ e $\beta=0$.

2.  O conjunto $\{(1,0,-1),(1,1,1),(1,1,0)\}\subset\mathbb{R}^3$ é L.I.. Com
	efeito, se $$\alpha(1,0,-1)+\beta(1,1,1)+\gamma(1,1,0)=(0,0,0)\,,$$
	então $$\begin{split}
	\alpha+\beta+\gamma&=0\\
	\beta+\gamma&=0\\
	-\alpha+\beta&=0\,.
	\end{split}$$ Vemos claramente que a lista $(0,0,0)$ é uma solução
	desse sistema. No entanto, se queremos provar que $X$ é L.I., devemos
	provar que o sistema tem uma única solução. Nesse caso em particular,
	podemos usar determinantes. O determinante do sistema é
	$$\begin{vmatrix}
	1&1&1\\
	0&1&1\\
	-1&1&0
	\end{vmatrix}=
	\begin{vmatrix}
	1&1&1\\
	0&1&1\\
	0&2&1
	\end{vmatrix}=-1\,.$$ Como ele é diferente de $0$, o sistema tem uma
	única solução.

Bases
=====

Diz-se que um conjunto $X$ em $\mathbb{R}^2$ (ou em $\mathbb{R}^3$) é um
**conjunto de geradores** se todo vetor de $\mathbb{R}^2$ (ou de
$\mathbb{R}^3$) pode ser escrito como uma combinação linear de vetores
de $X$.
